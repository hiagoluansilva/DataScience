**Preparação dos Dados e Estatísticas**

A biblioteca **Pandas** fornece estruturas de dados eficientes e fáceis de usar para trabalhar com tabelas e séries temporais. Ela foi criada para tornar a manipulação de dados em Python mais simples e intuitiva.

As duas principais estruturas da biblioteca são:

*   DataFrame → estrutura bidimensional (como uma planilha do Excel ou tabela de banco de dados).
*   Series → estrutura unidimensional (como uma coluna de tabela).


```python
import pandas as pd
```

```python
import numpy as np

# Carregar os dados
df = pd.read_csv('/content/dataset.txt', na_values=' ')
df
```

```python
df["Total"] = df["Quantidade"] * df["Preço"]
```

```python
# Exploração inicial dos dados
print(df.head())  # Visualizar as primeiras linhas do dataset
```

```python
print(df.info())  # Obter informações sobre tipos de dados e valores ausentes
```

```python
# Remoção de linhas duplicadas
df = df.drop_duplicates()
df
```

```python
# Remoção de linhas com valores ausentes
df = df.dropna()  # Remover linhas com valores ausentes
df
```

```python
# Remoção de linhas com quantidade menor que um
df = df[df['Quantidade'] >= 1]
df
```

```python
# Salvar novo dataset
df.to_csv('/content/datasetNEW.txt', index=False, sep=',')
```

```python
# Criando um array com valores e um NaN
dados = np.array([10, 20, 30, np.nan, 40, 50])

# Média ignorando valores NaN
print("Média:", np.nanmean(dados))
# Representa o valor médio dos dados, desconsiderando valores ausentes (NaN)

# Desvio padrão ignorando NaN
print("Desvio padrão:", np.nanstd(dados))
# Mede o quanto os valores estão dispersos em relação à média

# Variância ignorando NaN
print("Variância:", np.nanvar(dados))
# Mede a dispersão dos dados; é o quadrado do desvio padrão

# Mediana ignorando NaN
print("Mediana:", np.nanmedian(dados))
# Representa o valor central dos dados ordenados

# Soma ignorando NaN
print("Soma:", np.nansum(dados))
# Soma todos os valores numéricos, ignorando valores ausentes
```
